```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("gaming_addiction.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['subscription_status'].fillna(df['subscription_status'].mode()[0], inplace=True)
```

```python
df['addiction_severity'].fillna(df['addiction_severity'].mode()[0], inplace=True)
```

```python
df['depression_indicator'].fillna(df['depression_indicator'].mode()[0], inplace=True)
```

```python
df['gpa_or_performance_score'].fillna(
    df['gpa_or_performance_score'].median(),
    inplace=True
)
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
df = df.drop_duplicates()
df
```

```python
df.drop_duplicates(inplace=True)
df
```
